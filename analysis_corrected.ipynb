# Healthcare Cost Analysis

This project analyzes a healthcare insurance dataset to identify the factors that influence individual medical costs.

The workflow combines exploratory data analysis, statistical modeling, and machine learning to understand how demographic and lifestyle variables affect healthcare expenses.

**Exploratory data analysis was performed in Tableau**, while **Python was used for data preparation, regression modeling, residual analysis, and decision tree feature importance.**

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor

plt.rcParams['figure.figsize'] = (8, 5)
```

## Load Dataset

```python
df = pd.read_csv('insurance.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

## Data Preparation

Categorical variables such as smoking status and sex are encoded as numeric values so they can be used in regression and machine learning models.

```python
df_model = df.copy()
df_model['smoker'] = df_model['smoker'].map({'yes': 1, 'no': 0})
df_model['sex'] = df_model['sex'].map({'male': 1, 'female': 0})

df_model.head()
```

## 5. Exploratory Data Analysis

**Exploratory data analysis was conducted in Tableau**, not Python.

The Tableau visualizations created for this project include:

- Average healthcare cost by smoking status
- BMI vs healthcare cost scatter plot
- Age vs healthcare cost scatter plot
- Distribution of healthcare costs
- Average healthcare cost by region

These visuals helped identify the main patterns in the dataset before predictive modeling. The most important finding from the exploratory analysis was that **smoking status is the strongest driver of healthcare cost**, with **age** and **BMI** also showing meaningful relationships with charges.

## Linear Regression Model

```python
X = df_model[['age', 'bmi', 'children', 'smoker', 'sex']]
y = df_model['charges']

lin_reg = LinearRegression()
lin_reg.fit(X, y)

coefficients = pd.Series(lin_reg.coef_, index=X.columns)
coefficients
```

```python
coefficients.sort_values().plot(kind='barh')
plt.title('Regression Coefficients for Healthcare Cost Prediction')
plt.xlabel('Coefficient Value')
plt.tight_layout()
plt.show()
```

A linear regression model was used to estimate healthcare costs based on demographic and lifestyle variables. Smoking status had the largest effect on predicted healthcare costs, followed by age and BMI.

## Residual Analysis

```python
predictions = lin_reg.predict(X)
residuals = y - predictions

plt.scatter(predictions, residuals)
plt.axhline(0, linestyle='--')
plt.xlabel('Predicted Cost')
plt.ylabel('Residual')
plt.title('Residual Plot')
plt.tight_layout()
plt.show()
```

The residual plot helps evaluate model fit. The spread increases at higher predicted costs, suggesting that nonlinear relationships may also be present in the data.

## Decision Tree Model

```python
tree = DecisionTreeRegressor(random_state=42)
tree.fit(X, y)
```

## Feature Importance

```python
importances = pd.Series(tree.feature_importances_, index=X.columns).sort_values()

importances.plot(kind='barh')
plt.title('Decision Tree Feature Importance for Healthcare Cost Prediction')
plt.xlabel('Importance Score')
plt.tight_layout()
plt.show()
```

The decision tree model also identifies smoking status as the most important predictor of healthcare costs, with age and BMI contributing meaningfully to prediction.

## Key Findings

- Smoking status is the strongest driver of healthcare cost.
- Age and BMI also influence charges.
- Children have a smaller effect.
- Sex has relatively limited predictive value in the models used here.
- Using both Tableau and Python provides a clearer view of cost drivers through visualization and predictive modeling.
